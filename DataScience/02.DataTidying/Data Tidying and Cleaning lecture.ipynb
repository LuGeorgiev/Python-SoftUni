## Cleaning data

```python
weather_year = pd.read_csv("https://raw.githubusercontent.com/synesthesiam/blog/master/posts/data/weather_year.csv")
weather_year
```

```python
weather_year.columns = ["date", "max_temp", "mean_temp", "min_temp", "max_dew",
"mean_dew", "min_dew", "max_humidity", "mean_humidity",
"min_humidity", "max_pressure", "mean_pressure",
"min_pressure", "max_visibilty", "mean_visibility",
"min_visibility", "max_wind", "mean_wind", "min_wind",
"precipitation", "cloud_cover", "events", "wind_dir"]
```

```python
weather_year.shape
```

```python
def string_to_date(date_string):
    return datetime.datetime.strptime(date_string, "%Y-%m-%d")
```

```python
string_to_date("2012-3-10")
```

```python
weather_year.date = weather_year.date.apply(string_to_date)
```

```python
weather_year
```

```python
weather_year.precipitation.unique()
```

```python
weather_year.loc[weather_year.precipitation == 'T', "precipitation"] = 1e-5
```

```python
weather_year.precipitation = weather_year.precipitation.astype(float) 
weather_year
```

```python
weather_year.dtypes
```

```python
weather_year.events.unique()
```

```python
weather_year.describe().T
```
