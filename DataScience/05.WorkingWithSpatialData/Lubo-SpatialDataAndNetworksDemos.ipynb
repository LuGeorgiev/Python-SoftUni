```python

m = Basemap(
    projection = "merc", 
    llcrnrlat = -80, 
    urcrnrlat = 80,
    llcrnrlon = -180, 
    urcrnrlon = 180)
m.drawcoastlines()
m.fillcontinents(color = "coral", lake_color = "aqua")
m.drawparallels(np.arange(-90, 91, 30))
m.drawmeridians(np.arange(-180, 181, 60))
m.drawmapboundary(fill_color = "aqua")

m.plot(eartquake_x, earthquake_y, "o", color="red", markersize = 1)

plt.title("All earthquakes (M > 5.5)")
plt.show()
```

```python
volcanos_data = pd.read_csv("data/volcanos.csv")
volcanos_data.head()
```

```python

m = Basemap(
    projection = "merc", 
    llcrnrlat = -80, 
    urcrnrlat = 80,
    llcrnrlon = -180, 
    urcrnrlon = 180)
m.drawcoastlines()
m.fillcontinents(color = "coral", lake_color = "aqua")
m.drawparallels(np.arange(-90, 91, 30))
m.drawmeridians(np.arange(-180, 181, 60))
m.drawmapboundary(fill_color = "aqua")

volcano_x, volcano_y = m(volcanos_data.Longitude.tolist(), volcanos_data.Latitude.tolist())
m.plot(volcano_x, volcano_y, "x", color="green", markersize = 3)

plt.title("Major volcano eruptions")
plt.show()
```
