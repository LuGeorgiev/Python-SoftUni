```python
x = np.linspace(-2*np.pi, 2*np.pi, 5000)
```

```python
y = np.sin(x)
y_1 =np.cos(x)
plt.plot(x, y, label = "sin(x)")
plt.plot(x, y_1, label = "cos(x)")
plt.legend()
plt.show()
```

```python
def plot_math_functions(functions, x_min, x_max, num_point):
    x = np.linspace(x_min,x_max, num_point)
    for f in functions:
        y = f(x)    
        plt.plot(x, y)
    plt.gca().set_aspect("equal")
    plt.show()
```

```python
plot_math_functions([np.sin, np.cos, lambda x: 2*x -4], -2*np.pi, 2*np.pi, 5000)
```

#### draw cyrcle

```python
plot_math_functions ([lambda x:np.sqrt(1-x**2), lambda x: -np.sqrt(1-x**2)], -1, 1, 2000)
```

```python
import cmath
def solve_quadratic_equations(a, b, c):
    discriminant = cmath.sqrt(b**2 -4*a*c)
    return [(-b+discriminant)/(2*a),(-b-discriminant)/(2*a) ]
```

```python
print(solve_quadratic_equations(1,-3,-4))
print(solve_quadratic_equations(1,4,5))
```
