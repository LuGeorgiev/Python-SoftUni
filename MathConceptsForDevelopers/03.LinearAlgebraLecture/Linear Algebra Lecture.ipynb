## Linear Algebra

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt

import sympy  
```

```python
def plt_complex(numbers, colors, xlim = (-5, 5),ylim = (-5, 5)):
    plt.quiver(
        [0]*len(numbers),
        [0]*len(numbers),
        [number.real for number in numbers],
        [number.imag for number in numbers],
        scale = 1,
        scale_units = "xy",
        angles = "xy",
        color = colors)
    
    plt.gca().set_aspect("equal")
    plt.grid()
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.show()
```

```python
plt_complex([2+3j, 3 - 2j, 4], ["red", "blue", "grey"])
```

```python
x =sympy.symbols("x")
```

```python
sympy.solve((x-1)**2, x)
```

```python
solutions = sympy.solve((x-3)*(x-5)+40+x**3)
```

```python
sol = [complex(solution) for solution in solutions]
```

```python
plt_complex(sol, ["red", "blue"], (-8, 8), (-8,8))
```

```python
sol = sympy.solve(x**36+1)
solutions = [complex(solution) for solution in sol]
plt_complex(solutions, ["red", "blue", "green"], (-1.3, 1.3), (-1.3,1.3))
```
