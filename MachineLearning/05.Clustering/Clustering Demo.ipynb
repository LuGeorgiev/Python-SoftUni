```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs, make_moons

from sklearn.cluster import KMeans, AgglomerativeClustering

from scipy.cluster.hierarchy import linkage, dendrogram
```

# Cluestreing

## Live Demo

```python
blobs, blobs_clusters = make_blobs(n_samples=200)
```

```python
plt.scatter(blobs[:,0], blobs[:,1], c = blobs_clusters)
plt.show()
```

```python
k_means = KMeans(n_clusters=3, init = "random")
```

```python
k_means.fit(blobs)
```

```python
k_means.predict(blobs)
```

```python
predicted_blob_clusters = k_means.predict(blobs)
```

```python
plt.scatter(blobs[:,0], blobs[:,1], c = predicted_blob_clusters)
plt.show()
```

### defined cluster centers

```python
blobs, blobs_clusters = make_blobs(n_samples=200, centers = [(0, 0.5), (0.5, 0), (0, 0)])
```

```python
plt.scatter(blobs[:,0], blobs[:,1], c = blobs_clusters)
plt.show()
```

```python
k_means = KMeans(n_clusters=3, init = "random")
k_means.fit(blobs)
predicted_centered_blob_clusters = k_means.predict(blobs)
plt.scatter(blobs[:,0], blobs[:,1], c = predicted_centered_blob_clusters)
plt.show()
```

### defined standard deviation

```python
blobs, blobs_clusters = make_blobs(n_samples=200, cluster_std=2.4) # 1, 3, 5
plt.scatter(blobs[:,0], blobs[:,1], c = blobs_clusters)
plt.show()
```

```python
k_means = KMeans(n_clusters=3, init = "random")
k_means.fit(blobs)
predicted_deviated_blob_clusters = k_means.predict(blobs)
plt.scatter(blobs[:,0], blobs[:,1], c = predicted_deviated_blob_clusters)
plt.show()
```

### different deviation by X and Y axis

```python
blobs, blobs_clusters = make_blobs(n_samples=200, cluster_std=[(0.5, 5)] * 3)
plt.scatter(blobs[:,0], blobs[:,1], c = blobs_clusters)
plt.show()
```

```python
k_means = KMeans(n_clusters=3, init = "random")
k_means.fit(blobs)
predicted_deviated_blob_clusters = k_means.predict(blobs)

plt.gca().set_aspect("equal")
plt.scatter(blobs[:,0], blobs[:,1], c = predicted_deviated_blob_clusters)
plt.show()
```

# Moons

```python
moons, moon_clusters = make_moons(n_samples=200, noise = 0.07)

plt.scatter(moons[:,0], moons[:,1], c = moon_clusters)
plt.show()
```

```python
k_means = KMeans(n_clusters = 2, init = "random")
k_means.fit(moons)
predicted_moons = k_means.predict(moons)

plt.gca().set_aspect("equal")
plt.scatter(moons[:,0], moons[:,1], c = predicted_moons)
plt.show()
```

```python
k_means.cluster_centers_
```
